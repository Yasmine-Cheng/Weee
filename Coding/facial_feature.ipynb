```python
import numpy as np
import pandas as pd
import glob
```

```python
from google.colab import drive
drive.mount("/content/drive", force_remount=True)
```

```python
def facial_feature(dir:str):

    box = []
    filelocation = '/content/drive/Shareddrives/想聊對不隊/Code/FacialFeature/facial-features-'+dir+'/'

    for i in glob.glob(filelocation + '*.csv'): 
      df = pd.read_csv(i, index_col=None, header=0, engine='python')
      name = i.split('/')[-1].replace('.mp4.csv', '.npy')
      row = list(np.median(df, axis=0)[2:14])
      row.insert(0, name)
      box.append(row)

    return pd.DataFrame(box)
```

```python
test = facial_feature('test')
training = facial_feature('training')
validation = facial_feature('validation')
```

```python
df = pd.concat([test, training, validation], axis=0)
df = df.rename(columns={0: "index"})
df = df.dropna()
df.to_csv('/content/drive/Shareddrives/想聊對不隊/Code/CSV/all_facial_feature_result.csv', index=False)
df
```
